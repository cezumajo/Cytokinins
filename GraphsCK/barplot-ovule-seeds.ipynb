```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 4
ovule = (19, 10, 4, 8)
seeds = (1, 10, 16, 12)

ind = np.arange(N)   
width = 0.35  
```

```python
fig = plt.subplots(figsize =(5, 7))
p1 = plt.bar(ind, ovule, width)
p2 = plt.bar(ind, seeds, width,
             bottom = ovule)

plt.ylabel('total number')
plt.title('ovules and seeds observed at 3DAP')
plt.xticks(ind, ('Wildtype', 'ahk2', 'ahk3', 'ahk4'))
plt.yticks(np.arange(0, 30, 10))
plt.legend((p1[0], p2[0]), ('ovule', 'seeds'), bbox_to_anchor=(1.01, 1))


plt.savefig('Barplot-seedOvule',dpi=400, bbox_inches='tight')
```
